```python
#Encoding categorical column
df_copy['sex'] = df['sex'].map({'male':0,'female'1})
```

```python
df_copy
```

```python
#Feature Engineering
#Scle numerical feature(basic standardization)
df_copy['age_scaled'] = (df_copy['age'] - df_copy['age'].mean())/df_copy['age'].std()
```

```python
df_copy
```

```python
#Outlier handling : Remove extreme 'fare' values
q1 = df_copy['fare'].quantile(0.25)
q3 = df_copy['fare'].quantile(0.75)
iqr = q3-q1
upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr

df_copy = df_copy[(df_copy['fare'] < upper_limit) & (df_copy['fare'] > lower_limit)]
```
