```python
import random
```

```python
#random number generate garyo between 1 and 100
secret_number = random.randint(1, 100)
print(secret_number)

print("Welcome to the Number Guessing Game with Hints!")
print("Any number between 1 and 100.")

previous_guess = None  #ahile samma kei ni guess number rakheko xaina tesaile
```

```python

while True:
    guess = int(input("Enter your guess: "))

    #check gareko guess correct xa ki nai vanera
    if guess == secret_number:
        print("Correct! You guessed the number!")
        break
    else:
        if previous_guess is None:
            print("Wrong! Try again.")
        else:
# Compare the current guess with the previous guess to give feedback.
# If the current guess is closer to the secret number than the previous guess, print "Warmer"
# Otherwise print "Colder".
            if abs(secret_number - guess) < abs(secret_number - previous_guess):
                print("Warmer")
            else:
                print("Colder ")
        previous_guess=guess
```
