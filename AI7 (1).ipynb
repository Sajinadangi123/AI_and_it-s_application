```python
# Calculate the probability of drawing two aces and without replacement from a deck of 52 cards.
total_cards = 52
ace = 4
prob_ace = ace/total_cards
prob_first_ace = ace / total_cards
prob_second_ace = (ace-1)/(total_cards-1)
with_replacement = prob_first_ace * prob_second_ace
without_replacement = prob_ace * prob_ace
print("The prob of drawing two aces without replacement:",with_replacement)
print("The prob of drawing two aces with replacement:",without_replacement)
```

```python
# Task 2: Simulate rolling a six-sided die 50 times and calculate the experimental probability of rolling a 4. Print the result.
import random

trials = 50
count_4 = 0

for _ in range(trials):
    if random.randint(1, 6) == 4:
        count_4 += 1

probability = count_4 / trials
print("Experimental probability of rolling a 4 is:",probability)
```
