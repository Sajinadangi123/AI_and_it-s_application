DOUGHNUT CHART

```python
import matplotlib.pyplot as plt


labels = ['Roshmita', 'Sajina', 'Puja', 'Ram','Sita']
sizes = [10,30, 22, 25, 20]
colors = ['green','yellow', 'orange', 'red', 'blue']

# Doughnut chart
plt.pie(sizes, labels=labels, colors=colors, startangle=270, wedgeprops={'width': 0.5})
plt.axis('equal')  # circular banauna
plt.title('Name Distribution Doughnut Chart')
plt.show()
```
