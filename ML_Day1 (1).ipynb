# **Traditional Programming VS Machine Learning**

```python
# Traditional Programming: You give the rules and data, and get output
# Machine Learning: You give the data and output, and ML will find the rules

# Traditional Programming Example
def is_even(num):
    return num % 2 == 0

print(is_even(2))  # Output: True

# ML Thinking (Pseudocode)

# Data:
# X = [2, 4, 6, 8, 10]  # Even numbers
# Y = ['even', 'even', 'even', 'even', 'even']

# X = [1, 3, 5, 7, 9]   # Odd numbers
# Y = ['odd', 'odd', 'odd', 'odd', 'odd']

# Goal: Let the ML model learn patterns and classify new numbers like 11 or 12
```

# **Predicting Odd vs Even using Decision Tree**

```python
# Importing Libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  # fixed typo 'form' to 'from'

# Preparing Data
x = pd.DataFrame({'number': [1, 2, 3, 4, 5, 7, 8, 9, 10,11]})  # removed extra comma in [3,,4]
y = ['odd', 'even', 'odd', 'even', 'odd', 'odd', 'even', 'odd', 'even','odd']  # fixed to match x length

# Creating Model
model = DecisionTreeClassifier()

# Training model
model.fit(x, y)

# Making prediction
print(model.predict([[11]]))  # Should return 'even' or 'odd' depending on learned pattern
```

```python
# ML Example: Classifying Odd or Even Numbers
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# Step 1: Training data â€” both even and odd numbers
x = pd.DataFrame({'number': [-10, -5, -2, 0, 1, 2, 3, 4, 5, 6, 7, 1000]})
y = ['even', 'odd', 'even', 'even', 'odd', 'even', 'odd', 'even', 'odd', 'even', 'odd', 'even']

# Step 2: Train the model
model = DecisionTreeClassifier()
model.fit(x, y)

# Step 3: Test the model with new inputs
test_numbers = [[3], [100], [-5], [0], [13], [28]]

for num in test_numbers:
    prediction = model.predict([num])
    print(f"{num[0]} is predicted to be {prediction[0]}")
```
